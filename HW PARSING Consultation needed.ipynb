```python
import requests as rq
from bs4 import BeautifulSoup
import requests as rq
from bs4 import BeautifulSoup
url = 'https://theblueprint.ru' # сохраняем в переменную ссылку на страницу
page = rq.get(url)
html = page.text
soup = BeautifulSoup(page.text, 'html.parser')
print(soup.prettify())
links = [] # пустой список, будем сюда класть ссылки
souplist = soup.find_all('a', {'class': 'title'}, {'href' : True}) # находим элементы с этими параметрами, сохраняем в список
for elem in souplist: # для элемента списка
    links.append(elem['href'])
```

```python
print(links)
```

```python
adding = 'https://theblueprint.ru'
fulllinks = ['{0}{1}'.format(adding, j) for j in links]
```

```python
print(fulllinks)
```

```python
art_url = soup.find_all(fulllinks)
```
